```python
from itertools import product
from collections import Counter

x1 = np.linspace(1, 6, 6)
px1 = np.ones_like(x1) / 6

x2 = np.linspace(1, 6, 6)
px2 = np.ones_like(x2) / 6

c = Counter([sum(x) for x in product(x1, x2)])
y = list(c.keys())
py = np.array(list(c.values())) / np.sum([v for v in c])
py
```

```python
plt.subplot(311)
plt.stem(x1, px1)
plt.title("$X_1$ ")

plt.subplot(312)
plt.stem(x2, px2)
plt.title("$X_2$ ")

plt.subplot(313)
plt.stem(y, py)
plt.title("$Y=X_1+X_2$ ")

plt.tight_layout()
plt.show()
```
