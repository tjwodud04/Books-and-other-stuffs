```python
np.random.seed(15)
N = 7
data = np.sort(np.random.normal(size=(N)))[::-1]
```

```python
mean = np.mean(data)
mean
```

```python
plt.figure(figsize=(10, 2))
sns.rugplot(data, height=0.5, linewidth=4)
x = np.linspace(-3, 3, 100)
plt.axvline(x=0, ls=":", c="r", linewidth=2, label=" ")
plt.axvline(x=mean, ls="--", c="b", linewidth=2, label="")
plt.legend()
plt.xlim(-2, 2)
plt.show()
```

```python
distance_from_sample_mean = data - mean
distance_from_sample_mean
```

```python
sample_variance = (distance_from_sample_mean ** 2).mean()
sample_variance
```

```python
sample_variance * N / (N - 1)
```

```python
np.random.seed(0)
x1 = np.random.normal(size=100)
x2 = np.random.normal(size=100)
x = x1 + x2

print('X1  =', np.var(x1))
print('X2  =', np.var(x2))
print('X1 + X2  =', np.var(x))
```
